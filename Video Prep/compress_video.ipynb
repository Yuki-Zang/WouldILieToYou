##### `This file reduces the size of videos for the sake of annotation. All videos will be saved into a folder called "compressed video".`

Last updated on Sept. 24th, 2023.

```python
import os
import re
from string import Template
```

```python
path_to_dir = 'video' #path to where all videos are stored

try:
    os.mkdir('compressed video')
except:
    pass

for idir in os.listdir(path_to_dir):
    if idir != '.DS_Store' and not os.path.exists(f'compressed video/{idir}'):
        os.mkdir(f'compressed video/{idir}')

    ffmpeg = Template(f'ffmpeg -i \'$idir\'/\'$ifile\'  -vcodec libx264 -crf 30 -ar 16000 -ac 1 -vf scale=426:240 "compressed video"/"{idir}"/$ofile.mp4')
    if re.match(r'Series', idir):
        for ifile in os.listdir(path_to_dir + '/' + idir):
            ofile = re.search(r's\d+e\d+', ifile)[0]
            print(ofile)
            print(ifile, ofile, ffmpeg.substitute(idir=f"{path_to_dir}/{idir}", ifile=ifile, ofile=ofile))
            res = os.system(ffmpeg.substitute(idir=f"{path_to_dir}/{idir}", ifile=ifile, ofile=ofile))
```
